### Linear Regression
Linear regression is a type of supervised learning algorithm in machine learning that is used to establish a relationship between a dependent variable and one or more independent variables. In a linear regression algorithm, the goal is to find the line of best fit between the input features and the target variable, by choosing the values of the coefficients that minimize the distance between the predicted line and the training data points (Root Mean Square Error, or "RMSE"). The equation for linear regression can be represented as follows:

        Y = mx + b

Once the coefficients are determined, the model can be used to predict the values of the dependent variable for the new values of the independent variable(s). 

In our case, we'll run our regression using mileage as an input variable to predict price. It's reasonable to assume that a used car's mileage is an important predictor of its price: cars that have been used and worn down more will be less valuable, and cars that have been used less (i.e. closer to new) will be worth more. Most importantly, one can reason that a car's mileage is much more impactful on its value than other factors like its trim or color.

#### Ridge and lasso regression:
Introducing higher-order complexity, such as quadratic or cubic terms, into a model can cause the model to become overfitted. In order to combat this we ran the linear models through ridge and lasso regressions. Ridge regression prevents overfitting by reducing some of the coefficients for small but important features, while lasso regression identifies the most important features for a model's predictions by removing less important features. These methods eliminate unnecessary higher-order complexity, thus preventing the regression model from overfitting and making it simpler and easier to interpret.

##### For the trees and boosting models, we decided to run two of each model: one using only mileage to predict price, and one using mileage and year to predict price. 

### Trees
#### Decision trees:
A decision tree is a tree-like structure in which each internal node represents a test on an attribute, each branch represents the outcome of the rest, and each leaf node represents a class label or a numerical value. The basic idea behind decision trees is to iteratively partition the data into smaller and smaller subsets based on the value of a particular attribute. The attribute that is chosen to partition the data at each step is the one that provides the most information gain or accuracy. 

The effectiveness of decision trees depends on the quality of the attributes to split the data and the stopping criterion used to prevent overfitting. In general, decision trees can handle both categorical and continuous variables, can deal with missing values, and are easier for humans to interpret and visualize. However, decision trees can suffer from overfitting and may not always generalize well to new data, so it is important to choose appropriate stopping criteria. 
#### Random forests: 
Random forests are a popular ensemble learning method used in machine learning for classification and regression tasks. It is an extension of the decision tree algorithm that combines multiple decision trees to improve predictive accuracy and reduce overfitting.

Random Forests start by creating a set of decision trees. Each decision tree is created using a random subset of the data and a random subset of the features. By creating decision trees with different subsets of the data and features, random forests can reduce overfitting and improve the overall performance of the model. When making a prediction, each decision tree in the Random Forest makes a prediction, and the final prediction is made by taking the majority vote of all the decision trees. This technique is called bagging, which combines the output of multiple models to create a final model that has better performance than any individual model. 

Random Forests have several advantages over individual decision trees. First, Random Forests can handle non-linear relationships between the features and the target variable. Second, Random Forests can handle missing data, which can be problematic for decision trees. Third, Random Forests can handle unbalanced datasets, where one class has significantly fewer samples than another class. Compared to decision trees, Random Forests can improve the accuracy and robustness of the model. Decision trees can be prone to overfitting and can be sensitive to small changes in the training data. Random Forests can reduce overfitting by combining the output of multiple decision trees and can be more robust to changes in the training data because each decision tree is trained on a random subset of the data.

### Boosting
Boosting combines multiple weak models into a strong model to improve prediction accuracy. The goal of boosting is to sequentially train a set of weak models that can classify data with better accuracy than a single strong model.

The boosting process begins by training a base model on the training data. The base model is usually a simple model, such as a decision tree, that has a high bias and low variance. The base model makes predictions on the training data, and the predictions that are incorrect are given more weight. A second model is then trained on the same training data, but the weights of the incorrectly predicted data are increased. The process continues with more models being trained until a specified number of models are reached, or until a specific performance threshold is met.

The predictions of each model are then combined using a weighted average. The weights of each model are based on their performance, with more accurate models being given a higher weight. The final result is a strong model that is better at classifying data than any of the individual weak models.

Boosting has several advantages over other machine learning techniques. It can handle complex datasets with many features and can be used with a variety of machine learning algorithms. Boosting also has good generalization properties, which means it can predict well on new data.

### Neural Nets
Neural nets are inspired by the structure and function of the human brain, hence the term “neural.” Neural networks are composed of interconnected nodes or “neurons” that perform mathematical computations to transform input data into output data. 

The fundamental structure of a neural network consists of an input layer, one or more hidden layers, and an output layer. The input layer receives the raw data, which is then processed through the hidden layers, and finally, the output layer gives us our result. 

Each neuron in the neural network receives input from other neurons in the previous layer and produces an output that is fed into the next layer. Each neuron applies a mathematical function to its inputs and produces an output, which is then passed on to the next layer. The weights and biases of the connections between neurons are adjusted during the training process to optimize the performance of the neural network. 

Neural networks are widely used in machine learning because of their ability to learn complex relationships between inputs and outputs. They can be used for a variety of tasks such as image and speech recognition, natural language processing, and predictive modeling. One of the challenges with neural networks is determining the appropriate number of hidden layers and neurons in each layer. Additionally, neural nets can be computationally expensive and require large amounts of data to be efficient in training. 

For our neural nets, we decided to use 5 hidden layers, and run it two different times: one using mileage and year to predict price, and one using every feature (one-hot encoded) to predict price. We also ran the neural net using 2 and 3 layers as well, but eventually settled on using 5 layers as it produced the best results.
